```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0.0001, 1.57*2, 100)
Ki = 0.5 - 0.125 * (1 - np.cos(x)) + 0.75 * (1 - np.cos(x)) * np.log(0.5 * (1 - np.cos(x)))
x1=x*180/np.pi

plt.figure(figsize=(10,6))
plt.plot(x1, Ki, 'b-', label = "0.5-0.125*(1-np.cos(x))+0.75*(1-np.cos(x))*np.log(0.5*(1-np.cos(x)))")
#plt.xlim(-10, 190)
plt.title('Hellings & Downs curve')
plt.xlabel('Angle in degrees')
plt.ylabel('Correlation')
plt.grid(True)
plt.show()
```
